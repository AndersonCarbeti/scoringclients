# Mini notebook de test API — Pret a depenser

Ce notebook teste : `/predict`, `/predict-batch`, `/predict-by-id/{client_id}`.

```python
# Config
BASE_URL = 'http://localhost:8000'
SAMPLE_PREDICT = 'sample_predict.json'
SAMPLE_BATCH = 'sample_batch.json'
CLIENT_ID = 100001
```

```python
import json
import requests
```

```python
# Health
requests.get(f'{BASE_URL}/health').json()
```

```python
# Model info
requests.get(f'{BASE_URL}/model-info').json()
```

```python
# Predict (client unique)
payload = json.load(open(SAMPLE_PREDICT))
requests.post(f'{BASE_URL}/predict', json=payload).json()
```

```python
# Predict batch
payload = json.load(open(SAMPLE_BATCH))
requests.post(f'{BASE_URL}/predict-batch', json=payload).json()
```

```python
# Predict by id
requests.get(f'{BASE_URL}/predict-by-id/{CLIENT_ID}').json()
```
